```python
import numpy as np
a = np.arange(-10, 10, 0.1)
```

```python
a_tanh = np.tanh(a)
```

```python
def sigmoid(x):
    return np.where(x>0, 1/(1-np.exp(-x)), 0)
a_sig = sigmoid(a)
```

```python
def relu(x):
    return np.where(x>=0, x, 0)
a_relu = relu(a)
```

```python
def leakyRelu(x):
    return np.where(x>=0, x, 0.01*x)

a_leaky_relu = leakyRelu(a)
```

```python
def softmax(x):
    e_x = np.exp(x-np.max(x))
    return e_x/np.sum(e_x)
a_softmax = softmax(a)
```

```python
import matplotlib.pyplot as plt
alpha = 0.7
plt.figure(figsize=(10, 6))
plt.plot(a, a_softmax, color='red', label='softmax', alpha=alpha)
plt.plot(a, a_tanh, color='blue', label='tanh', alpha=alpha)
plt.plot(a, a_relu, color='orange', label='ReLU', alpha=1)
plt.plot(a, a_leaky_relu, color='green', linestyle='--', label='Leaky ReLU', alpha=1)
plt.fill_between(a, a_sig, color='yellow', label='Sigmoid', alpha=0.5)
plt.legend()
plt.show()
```
